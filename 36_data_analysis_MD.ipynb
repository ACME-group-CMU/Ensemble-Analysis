```python
import numpy as np
import matplotlib.pyplot as plt
import os
import tqdm

# Import v2 modules with abbreviated names
import src.data_management_v2 as dm
from src.structure_manager_v2 import StructureManager
import src.rdf_v2 as rdf
import src.ensembles_v2 as ens
import src.Birch_Murnaghan2 as bm
import src.entropy as ent
import quests

# Set random seed for reproducibility
np.random.seed(42)

folder_path = "data/all_sizes"
vasp24 = [f for f in os.listdir(folder_path) if f.startswith('SiO2_24')]
vasp36 = [f for f in os.listdir(folder_path) if f.startswith('SiO2_36')]
vasp48 = [f for f in os.listdir(folder_path) if f.startswith('SiO2_48')]
vaspall = [f for f in os.listdir(folder_path) if f.startswith('SiO2')]
struct_ids = [f.replace('.vasp', '') for f in vaspall]
struct_ids_24 = [f.replace('.vasp', '') for f in vasp24]
struct_ids_36 = [f.replace('.vasp', '') for f in vasp36]
struct_ids_48 = [f.replace('.vasp', '') for f in vasp48]

struct_ids_36.sort()
struct_ids_24.sort()
struct_ids_48.sort()
struct_ids.sort()   
```

```python
dm.populate_sq(struct_ids, folder_path, weighting='neutron')
dm.populate_gr_from_sq(struct_ids, folder_path, weighting='neutron')
```

```python
sq_data = dm.load_sq(struct_ids_24, pairs='total')
ens = rdf.calculate_ensemble_sq(struct_ids_24, sq_data=sq_data)
q = ens['q']

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(q, q*(ens['S_total']-1), lw=2, label=f'Ensemble ({len(struct_ids_24)})')
ax.axhline(0, color='k', lw=0.5, ls='--')
ax.set_xlabel('q [Å⁻¹]')
ax.set_ylabel('q(S(q)−1) [Å⁻¹]')
ax.set_title('Ensemble S(q) — 24-atom, Neutron')
ax.set_ylim(-3, 3)
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Ensemble q(S-1) max: {(q*(ens['S_total']-1)).max():.3f}")
print(f"Ensemble q(S-1) min: {(q*(ens['S_total']-1)).min():.3f}")
```

```python
sq_data = dm.load_sq(struct_ids_24, pairs='all')
gr_data = dm.load_gr_from_sq(struct_ids_24, pairs='all')
rdfs = dm.load_rdfs(struct_ids_24, pairs='all', smoothed=True)
densities = dm.load_densities(struct_ids_24)
energies = dm.load_energies(struct_ids_24)

ens_sq = rdf.calculate_ensemble_sq(struct_ids_24, sq_data=sq_data)
ens_gr = rdf.calculate_ensemble_gr_from_sq(struct_ids_24, gr_sq_data=gr_data)
ens_direct = rdf.calculate_ensemble_partial_rdfs(struct_ids_24, rdfs=rdfs, 
                                                  energies=energies, densities=densities)

fig, axes = plt.subplots(3, 1, figsize=(8, 10))
for ax, pair in zip(axes, [('Si','Si'), ('Si','O'), ('O','O')]):
    r_ft = ens_gr['r']
    g_ft = ens_gr['g_partials'][pair]
    r_d, g_d = ens_direct[pair]
    
    n_beta = ens_sq['partial_densities_abs'][pair[1]]
    
    ax.plot(r_d,  n_beta * 4*np.pi*r_d**2  * g_d,  color='orange', label='Direct (smoothed)')
    ax.plot(r_ft, n_beta * 4*np.pi*r_ft**2 * g_ft, color='red',    label='FT from S(q)')
    ax.set_title(f'{pair[0]}-{pair[1]}')
    ax.set_xlabel('r [Å]')
    ax.set_ylabel(r'$n_\beta 4\pi r^2 g_{\alpha\beta}(r)$')
    ax.set_xlim(0.5, 6)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Test: assemble S_total using Eq. B7 instead of Eq. 3
# S_AL_ab = (1/sqrt(N_a*N_b)) * sum sinc  (no concentration factor)
# S_total = sum_ab sqrt(c_a*c_b)*f_a*f_b/<f>^2 * [S_AL_ab - delta_ab] + 1

structure, _ = dm.vasp_to_pymatgen(struct_ids_24[0], folder_path)
sq = rdf.calculate_sq(structure, weighting='neutron')
q = sq['q']
c = sq['concentrations']
b = rdf.NEUTRON_B
f_mean = sum(c[e]*b[e] for e in c)
f_mean_sq = f_mean**2
elements = list(c.keys())

# Current S_total (Eq. 3 with c_a*N_a norm on same-species)
S_current = sq['S_total']

# What if same-species used N_a norm (original) and assembly used sqrt(c_a*c_b)?
# i.e. treat all partials as S_AL and use B7 weights
print("Current S_total range:", S_current.min().round(3), S_current.max().round(3))
print()
print("Weights comparison:")
for e1 in elements:
    for e2 in elements:
        w_eq3 = c[e1]*c[e2]*b[e1]*b[e2]/f_mean_sq
        w_b7  = np.sqrt(c[e1]*c[e2])*b[e1]*b[e2]/f_mean_sq
        print(f"  {e1}-{e2}: Eq3 w={w_eq3:.4f}, B7 w={w_b7:.4f}")
```
